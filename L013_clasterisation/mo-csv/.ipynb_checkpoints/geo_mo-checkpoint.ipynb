```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import path
```

```python
#читаем и готовим данные о координатах полигонов (мун.округов Мск)
with open("mo.geojson", "r") as read_file:
    geo2 = json.load(read_file)
```

```python
#преобразуем структуру данных в словарь для, в свою очередь, преобразования его в DataFrame
dict_ = {}
key_ = 0
for i, district in enumerate(geo2['features']):
    for j, polygon in enumerate(district['geometry']['coordinates']):       
        if district['geometry']['type'] == 'MultiPolygon':
            polygon_ = sum(polygon, [])
        else:
            polygon_ = polygon
        
        dict_[key_] = \
                {   
                    'NAME' : district['properties']['NAME'],
                    'TYPE_MO' : district['properties']['TYPE_MO'],
                    'OKATO' : district['properties']['OKATO'],
                    'OKTMO' : district['properties']['OKTMO'],
                    'NAME_AO' : district['properties']['NAME_AO'],
                    'OKATO_AO' : district['properties']['OKATO_AO'],
                    'ABBREV_AO' : district['properties']['ABBREV_AO'],
                    'POLYGON_TYPE' : district['geometry']['type'],
                    'POLYGON' : polygon_
                }
        key_ += 1
```

```python
#получим DataFrame
df_geo = pd.DataFrame(pd.DataFrame(dict_).T)
```

```python
#вытащим координаты всех полигонов и нарисуем полигоны, если надо
#тут будут храниться готовые к рисованию координаты всех полигонов
geo_ = []

def geo_append(row, draw = False):
    geo_x = []
    geo_y = []
    
    for point in row['POLYGON']:
        geo_x.append(point[0])
        geo_y.append(point[1])
    
    geo_.append([geo_x, geo_y])
    
    if draw:
        plt.plot(geo_x, geo_y)
        plt.title(row['NAME'] + ', ' + row['NAME_AO'] + ' АО')
        plt.show()

list_ = [df_geo.apply(geo_append, axis = 1, draw = False)]
```

```python
#читаем данные об обращениях жителей
df = pd.read_excel('geo_comment.xlsx', sheet_name = 'Sheet1', index_col = 0, header = 0) #, usecols = 'A:E')
```

```python
#сопоставим каждой точке (сообщению) муниципальный округ
#через принадлежность координат точки одному из полигонов

#преобразуем координаты в списки кортежей (по числу полигонов)
geo_tuple = [[(polygon[0][point], polygon[1][point]) for point in range(len(polygon[0]))] for polygon in geo_]
```

```python
%%time
#перебор сообщений, проверка принадлежности сообщения полигону, заполнение признака "mo"
def message_to_polygon(row, geo_tuple):
    for index, polygon_xy in enumerate(geo_tuple):
        if path.Path(polygon_xy).contains_point((row['x'], row['y'])):
            return index
    return -1

df['mo'] = df.apply(message_to_polygon, geo_tuple = geo_tuple, axis = 1)
```

```python
#сохраним DF с новым признаком "mo" (муниципальный округ)
df = pd.read_csv('geo_comment_mo.zip', compression = 'zip')
```

```python
#найдем количество проголосовавших в разрезе мун.округов
df_group_mo = pd.DataFrame(df.groupby('mo')['multiplier'].sum().reset_index())
```

```python
df_group_mo
```

```python
#нарисуем все полигоны на одном листе
plt.figure(figsize=(50,50))

#рисуем полигоны
for polygon in geo_:
    plt.plot(polygon[0], polygon[1])

#рисуем в центре полигона количество проголосовавших
def text_print(row):
    #находим центр полигона
    x_ = 0
    y_ = 0
    point_count = len(geo_[row['mo']][0])
    for count in range(0, point_count):
        x_ += geo_[row['mo']][0][count]
        y_ += geo_[row['mo']][1][count]
    #рисуем текст
    plt.annotate(row['multiplier'],(x_ / point_count, y_ / point_count))
    
df_group_mo[df_group_mo['mo'] != -1].apply(text_print, axis = 1)

#plt.scatter(df['x'], df['y'])

plt.show()
```

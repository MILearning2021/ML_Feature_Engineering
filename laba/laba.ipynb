```python
#обработчик вариантов с/без PCA, модель - лес с подбором количества деревьев, и др.параметры
def rfc_pca(X, Y, pca = None, pca_components = None, n_best = 10, over = None, k_fold = None):
    #делим датасет
    X_tr, X_t, Y_tr, Y_t = train_test_split(X, Y, test_size = 0.3, shuffle = True, stratify = Y, \
                                            random_state = 42)
    
    #балансируем тренировочную выборку по целевой переменной, если это нужно
    if over == 'SMOTE':
        X_tr, Y_tr = SMOTE(sampling_strategy = 'not majority', k_neighbors = 2, \
                           random_state = 42).fit_resample(X_tr, Y_tr)
    elif over == 'ADASYN':
        X_tr, Y_tr = ADASYN(sampling_strategy = 'not majority', n_neighbors = 2, \
                            random_state = 42).fit_resample(X_tr, Y_tr)
    elif over == 'ROS':
        X_tr, Y_tr = RandomOverSampler(random_state = 42).fit_resample(X_tr, Y_tr)
        
    #принт свойств тренировочного набора
    if over != None:
        print(f'Размерность тренировочного набора: {X_tr.shape}')
        print(f'Баланс классов тренировочного набора:\n{Y_tr.value_counts(normalize = True)}')
    
    #с/без PCA    
    if pca == True:
        #стандартизация признаков
        scal = StandardScaler().fit(X_tr)
        #нахождение главных компонент
        pca = PCA(n_components = pca_components).fit(scal.transform(X_tr))
        x_tr = pca.transform(scal.transform(X_tr))
        x_t = pca.transform(scal.transform(X_t))
        print(f'Количество главных компонент: {pca_components}')
        print(f'Доля объясненной дисперсии целевой переменной: {pca.explained_variance_ratio_.sum()}')
    else:
        x_tr = X_tr
        x_t = X_t
    
    #лес с подбором количества деревьев   
    parameters = {'n_estimators': np.linspace(10, 230, 50, dtype = np.int),
                  'random_state' : [42]}
    clf = GridSearchCV(RandomForestClassifier(), parameters)
    clf.fit(x_tr, Y_tr)
    
    #кросс-валидация
    if k_fold != None:
        scores = cross_validate(clf.best_estimator_, X_tr, Y_tr, scoring = 'roc_auc_ovr', \
                                cv = k_fold, return_estimator = True)
        print(f'\nРезультаты кросс-валидации "roc_auc_ovr": {scores["test_score"]}\n')
        
    #предсказание лучшей модели и принт метрики качества
    print(f'ROC AUC на тесте: {roc_auc_score(Y_t, clf.best_estimator_.predict_proba(x_t), multi_class = "ovr")}') 
   
    #важность признаков
    if pca == None: 
        feature_imp(n_best = n_best, rfr = clf.best_estimator_, X = x_t)
    
    pass

#важность признаков
def feature_imp(rfr, X, n_best = 10):
    imp_sort = []
    feature_sort = []

    col_imp = [[X.columns[i], rfr.feature_importances_[i]] for i in range(0, len(X.columns))]
    col_imp.sort(key = lambda x: x[1], reverse = True)
    for item in col_imp[:n_best]:
        imp_sort.append(item[1]); 
        feature_sort.append(item[0])

    plt.barh(np.arange(n_best), imp_sort)
    plt.yticks(np.arange(n_best), feature_sort)
    plt.title('Важность признаков')
    plt.xlabel('важность')
    plt.ylabel('признак')
    plt.show()
    
    pass
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, MinMaxScaler
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import r2_score
```

```python
df_scaled = StandardScaler().fit_transform(PolynomialFeatures(degree = 3).fit_transform(df))
X_tr, X_t, Y_tr, Y_t = train_test_split(df_scaled, y, test_size = 0.2, random_state = 42)
```

```python
res = {}

def execute_RSCV(est, params, cv = 7, n_iter = 20):
    rs = RandomizedSearchCV(estimator = est, param_distributions = params , cv = cv, n_iter = n_iter)
    rs.fit(X_tr, Y_tr)

    res[est] = {
                    'best_score_' : rs.best_score_,
                    'best_estimator_' : rs.best_estimator_,
                    'best_params_' : rs.best_params_,
                    'r2_test' : r2_score(rs.best_estimator_.predict(X_t), Y_t)
            }
    
    print(f'------------------\nЛучший R^2 на трейне: {res[est]["best_score_"]}')
    print(f'Параметры лучшего эстиматора:\n{res[est]["best_params_"]}')
    print(f'R^2 на тесте: {res[est]["r2_test"]}\n------------------')
    
    pass
```

```python
import multiprocessing
n_jobs = multiprocessing.cpu_count() // 2
```

```python
%%time
#В1 - ElasticNet

params = {
            'l1_ratio' : np.linspace(0.01, .99, 100),
            'alpha' : np.linspace(0.01, .99, 100)
        }

execute_RSCV(est = ElasticNet(max_iter = 10000, random_state = 42), params = params , cv = 7, n_iter = 20)
```

```python
%%time
#В2 - RandomForestRegressor

params = {
            'n_estimators' : np.linspace(100, 700, 100, dtype = 'int'),
            'max_depth' : np.linspace(3, 16, 10, dtype = 'int'),
            'min_samples_leaf' : np.linspace(1, 6, 3, dtype = 'int')
        }

execute_RSCV(est = RandomForestRegressor(n_jobs = n_jobs, random_state = 42), params = params , cv = 3, n_iter = 10)
```

```python
%%time
#В3 - XGBRegressor

params = {
            'n_estimators' : np.linspace(100, 700, 100, dtype = 'int'),
            'max_depth' : np.linspace(3, 16, 10, dtype = 'int'),
            'gamma' : np.linspace(1., 100., 100),
            'lambda' : np.linspace(1., 100., 100),
            'alpha' : np.linspace(0.01, 1., 100),
            'eta' : np.linspace(0.01, 1., 10)
        }

execute_RSCV(est = XGBRegressor(n_jobs = n_jobs, random_state = 42), params = params , cv = 3, n_iter = 10)
```

```python
#В4 - нейронная сеть

#используем стандартизированные значения X и Y (использую MinMaxScaler, StandardScaler не зашел)
df_scaled = MinMaxScaler().fit_transform(PolynomialFeatures(degree = 3).fit_transform(df))
y_scaled = MinMaxScaler().fit_transform(y.reshape(-1, 1))
#получим трейн и тест
X_tr, X_t, Y_tr, Y_t = train_test_split(df_scaled, y_scaled, test_size = 0.2, random_state = 42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
#строим сеть с одним внутренним слоем
def my_model_execute(epochs):
    #сеть-фантазия
    model = Sequential()
    model.add(Dense(128, input_dim = X_tr.shape[1], activation = 'relu'))
    #model.add(Dropout(0.1))
    model.add(Dense(506, activation = 'relu'))
    #model.add(Dropout(0.3))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(optimizer = 'adam', loss = 'mae', metrics = ['mae'])
    #тренировка модели с использованием кросс-валидации
    history = model.fit(X_tr, Y_tr, validation_split = .2, batch_size = 16, epochs = epochs, \
                    shuffle = True, verbose = 0)
    #возвращаем историю валидации, обученную модель
    return history, model
```

```python
%%time
#тренируем и сохраняем историю точность/потери
epochs = 140
hist, model = my_model_execute(epochs)
```

```python
#графики
acc = hist.history['mae']
val_acc = hist.history['val_mae']

epochs_range = range(epochs)

plt.figure(figsize=(10,5))

plt.plot(epochs_range, acc, label = 'MAE на обучении')
plt.plot(epochs_range, val_acc, label = 'MAE на валидации')
plt.legend(loc = 'upper right')
plt.title('Точность на обучающих и валидационных данных')

plt.show()
```

```python
print(f'------------------\nR^2 на трейне: {r2_score(model.predict(X_tr), Y_tr)}')
print(f'R^2 на тесте: {r2_score(model.predict(X_t), Y_t)}\n------------------')
```

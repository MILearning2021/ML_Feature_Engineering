```python
#формирую набор для тренировки: пропущенные значения в shot_made_flag определяют тестовый набор
df = data[data['shot_made_flag'].isna() == False].copy()
df['shot_made_flag'] = df['shot_made_flag'].astype('int32')
target = df['shot_made_flag']
df.drop(['shot_made_flag'], axis = 1, inplace = True)
#формирую тестовый набор для kaggle, итого 5000 тестовых сэмплов
df_kaggle = data[data['shot_made_flag'].isna()].copy()
```

```python
#категориальные признаки
#удалить признак - 'matchup', 'team_name', 'team_id', 'shot_id', 'game_event_id'
#признаки 'game_date', 'season' - как их закодировать? будем использовать, как есть
#признаки 'combined_shot_type', 'action_type' - combined_shot_type аггрегат от action_type
#признаки 'playoffs', 'period', 'game_id' - перекодировать в категориальные
#сформировать новый признак 'sec_remaining' = 60 * 'minutes_remaining' + 'seconds_remaining'
#неадекватных выбросов не обнаружил
```

```python
#не понятные переменные 'loc_x' и 'loc_y' - проверить на важность
```

```python
#тут будут функции для пайплайна
#удаляю ненужные признаки
def feature_drop(dataframe, features):
    dataframe.drop(features, axis = 1, inplace = True)
    pass
#кодирую признаки в категориальные для dummy преобразования 
def feature_coding(dataframe, features):
    for feature in features:
        dataframe[feature] = dataframe[feature].astype('object')
    pass
#формирую новый признак sec_remaining
def new_future(row):
    return (60 * row['minutes_remaining']) + row['seconds_remaining']
#пайплайн подготовки признаков
def feature_prep(dataframe, feat_to_drop, feat_to_coding):
    #удаляю ненужные признаки
    feature_drop(dataframe, feat_to_drop)
    #кодирую признаки в категориальные
    feature_coding(dataframe, feat_to_coding)
    #формирую новый признак sec_remaining
    dataframe['sec_remaining'] = dataframe.apply(new_future, axis = 1)
    feature_drop(dataframe, ['minutes_remaining', 'seconds_remaining'])
    #dummy кодирование
    #return StandardScaler().fit_transform(pd.get_dummies(dataframe, drop_first = False))
    return pd.get_dummies(dataframe, drop_first = True)
```

```python
#работаю с набором трейн
#feat_to_drop = ['matchup', 'team_name', 'team_id', 'shot_id', 'game_event_id', 'game_date', 'game_id']
feat_to_drop = ['matchup', 'team_name', 'team_id', 'shot_id', 'game_event_id', 'combined_shot_type', \
                'game_date', 'game_id']
#feat_to_drop = ['matchup', 'team_name', 'team_id', 'shot_id', 'game_event_id', 'action_type', \
                #'game_date', 'game_id']
#feat_to_coding = ['playoffs', 'period', 'game_id']
feat_to_coding = ['playoffs', 'period']
#подготовка признаков
df_dummy_sc = feature_prep(df, feat_to_drop, feat_to_coding)
```

```python
df_dummy_sc.shape
```

```python
#модели
models = {'LinearRegr' : LinearRegression(), 'Lasso' : Lasso(alpha=0.0001), 'Ridge' : Ridge(alpha=0.1)}
```

```python
X_tr, X_t, Y_tr, Y_t = train_test_split(df_dummy_sc, target, test_size = 0.2, shuffle = True, \
                                        stratify = target, random_state = 42)
```

```python
%%time
def train_test_models(models, X_tr, X_t, Y_tr, Y_t):
    models_res = {}
    for model in models:
        preds = []
        models[model].fit(X_tr, Y_tr)
        for pred in models[model].predict(X_t):
            if pred > 0.5: 
                preds.append(1)
            else: 
                preds.append(0)
        print(f'{model}: acc {accuracy_score(Y_t, preds)}, roc_auc {roc_auc_score(Y_t, models[model].predict(X_t))}')
        models_res[model] = {
                                'model' : models[model],
                                'accuracy' : accuracy_score(Y_t, preds),
                                'preds' : list(preds),
                                'coef' : models[model].coef_
                            }
    return models_res

m_r = train_test_models(models, X_tr, X_t, Y_tr, Y_t)
```

```python
coefs = pd.DataFrame({'Lasso' : m_r['Lasso']['coef'], 
                      'Ridge' : m_r['Ridge']['coef'],
                      'LinearRegr' : m_r['LinearRegr']['coef']})
```

```python
plt.figure(figsize = (12,25))
plt.subplot(1,2,1)
plt.plot(coefs['Lasso'], X_tr.columns, label = 'Lasso')
plt.plot(coefs['Ridge'], X_tr.columns, label = 'Ridge')
plt.legend(loc='upper left')
plt.subplot(1,2,2)
plt.plot(coefs['LinearRegr'], coefs.index, label = 'LinearRegr')
plt.legend(loc='upper right')
```

```python
#модели Ridge и Lasso, в целом, похоже назначают коэффициенты признакам: обе модели выделили существенное
#влияние на целевую переменную признаков, обозначающих тип броска, и признаков, обозначающих зону площадки, 
#откуда производился бросок; влияние признаков игрового сезона не великож; влияние того  обстоятельства, кто 
#является соперником, фрагментарно
```

```python
#модель LinearRegression, внезапно, придала огромный вес признаку 'lon' (longitude?) - не понятный признак;
#в остальном, если рассматривать график коэффициентов LR по частям, убирая признаки с высоким значением коэффициентов,
#получается, что LR, в целом, отследила те же тенденции влияния признаков на целевую переменную, как и две
#другие модели; замеченная особенность - у LR меньше 0-х коэффициентов
```

```python
#PCA разложение
pca = PCA(n_components = 5).fit(X_tr)
print(pca.fit(X_tr).explained_variance_ratio_)
```

```python
#видим, что первые 3 компоненты совместно объясняют 99.947% дисперсии целевой переменной
```

```python
#вот самая "влиятельная" компонента, для примера
print(pca.fit(X_tr).components_[0])
```

```python
#проведем эксперимент: обучим наши 3 модели на "главных компонентах"
#В1 - берем 5 самых главных компонент
pca = PCA(n_components = 5).fit(X_tr)
x_tr = pca.transform(X_tr)
x_t = pca.transform(X_t)
m_r_pca = train_test_models(models, x_tr, x_t, Y_tr, Y_t)
```

```python
#результат работы моделей значительно ухудшился
```

```python
#В2 - берем ВСЕ главные компоненты
pca = PCA(n_components = None).fit(X_tr)
x_tr = pca.transform(X_tr)
x_t = pca.transform(X_t)
m_r_pca = train_test_models(models, x_tr, x_t, Y_tr, Y_t)
```

```python
#результат работы моделей вернулся к прежним показателям (улучшился по сравнению с вариантом для 5 компонент)
```
